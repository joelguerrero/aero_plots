```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
#For minor adn major ticks
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
```

```python
#Read in data from XFOIL
data=np.loadtxt('./my_outputs/63415_v1.dat', skiprows=12)
```

```python
#Plot extended polar

#fig = plt.figure()
fig, ax = plt.subplots()

fig.set_size_inches(12,12)

twin1 = ax.twinx()
twin2 = ax.twinx()
twin3 = ax.twiny()



#fig.subplots_adjust(right=0.75)
#fig.subplots_adjust(bottom=0.2)



# Offset the right spine of twin2.  The ticks and label have already been
# placed on the right by twinx above.
twin2.spines["right"].set_position(("axes", 1.2))


#Second bottom axis
twin3.xaxis.set_ticks_position('bottom')
twin3.xaxis.set_label_position('bottom')
twin3.spines['bottom'].set_position(('axes', -0.15))



#Axis defined from this point on
#ax left vertical axis
#twin1 right vertical axis
#twin2 second right  vertical axis
#twin3 second bottom axis



#Plot curves in each axes
p1, = ax.plot(data[:,0], data[:,1], '-o',color='blue', linewidth=1, zorder=101, 
              label='Lift coefficient $c_l$')
p2, = twin1.plot(data[:,0], data[:,2], '-o',color='red', linewidth=1, zorder=102, 
                 label='Drag coefficient $c_d$')
p3, = twin2.plot(data[:,0], data[:,4], '-o',color='green', linewidth=1, zorder=103, 
                 label='Moment coefficient $c_m$')
p4, = twin3.plot(data[:,2], data[:,1], '-o',color='orange', linewidth=1, zorder=104, 
                 label='Polar plot $c_d \ vs. \ c_l$')
#p4, = plt.plot(data[:,2], data[:,1], '-o',color='orange', label='Polar plot')



#Axes limits
ax.set_xlim(-15, 30)

ax.set_ylim(-1, 3)
twin1.set_ylim(0, 0.4)
twin2.set_ylim(-0.5, 0.5)

twin3.set_xlim(0, 0.4)
#twin3.set_ylim(-1, 2)



#Axes labels
#ax.set_xlabel('AOA ($^\circ$)', fontsize=18)
ax.set_xlabel('AOA ($^\circ$)',fontsize=16)
ax.set_ylabel('Lift coefficient $c_l$',fontsize=16)
twin1.set_ylabel('Drag coefficient $c_d$',fontsize=16)
twin2.set_ylabel('Moment coefficient $c_m$',fontsize=16)
twin3.set_xlabel('Drag coefficient $c_d$',fontsize=16)



#Get axes colors
ax.yaxis.label.set_color(p1.get_color())
twin1.yaxis.label.set_color(p2.get_color())
twin2.yaxis.label.set_color(p3.get_color())
twin3.xaxis.label.set_color(p4.get_color())



#Axes ticks format
tkw = dict(size=10, width=2, labelsize=16)
ax.tick_params(axis='y', colors=p1.get_color(), **tkw)
twin1.tick_params(axis='y', colors=p2.get_color(), **tkw)
twin2.tick_params(axis='y', colors=p3.get_color(), **tkw)
twin3.tick_params(axis='x', colors=p4.get_color(), **tkw)
ax.tick_params(axis='x', **tkw)



#Legend
ax.legend(loc=2,handles=[p1, p2, p3, p4],fontsize=12)



#Axes minor ticks
ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.tick_params(which='minor', size=5, width=1)

ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.tick_params(axis='y',which='minor', size=5, width=1,colors=p1.get_color())
#ax.tick_params(axis='y', which='both', colors='red')

twin1.yaxis.set_minor_locator(AutoMinorLocator())
twin1.tick_params(which='minor', size=5, width=1,colors=p2.get_color())

twin2.yaxis.set_minor_locator(AutoMinorLocator())
twin2.tick_params(which='minor', size=5, width=1,colors=p3.get_color())

twin3.xaxis.set_minor_locator(AutoMinorLocator())
twin3.tick_params(which='minor', size=5, width=1,colors=p4.get_color())
ax.tick_params(axis='y', colors=p1.get_color(), **tkw)



#Plot grid in reference to original axes x-y
#ax.grid()
ax.grid(which='major', linestyle='-')
ax.grid(which='minor', linestyle='-.', linewidth=0.5)
#ax.grid(b=True, which='minor', color='r', linestyle='--')



#Save figure
#plt.savefig('fig1.png', dpi=300)
#plt.savefig('fig2.png', dpi=300, bbox_inches='tight')


plt.show()
```
