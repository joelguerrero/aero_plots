```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
#For minor adn major ticks
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
```

```python
#Read in data from XFOIL
data=np.loadtxt('./my_outputs/63415_v1.dat', skiprows=12)

#Read cp data
data1=np.loadtxt('./my_outputs/63415_v1.cp', skiprows=3)

#Read cp data for top and bottom surfaces
data2=np.loadtxt('./my_outputs/63415_v1_top.cp', skiprows=3)
data3=np.loadtxt('./my_outputs/63415_v1_bottom.cp', skiprows=3)
```

```python
#Read in data from XFOIL
airfoil=np.loadtxt('./my_airfoils/63415v1.dat', skiprows=1)
```

```python
fig = plt.figure()
ax = fig.add_subplot()


#Figure size
fig.set_size_inches(12,6)


#Plot airfoil data
plt.plot(airfoil[:,0],airfoil[:,1],
         linewidth=1,color='black')


#Axes label
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)


#Bounds, grid, ticks fontsize
#plt.xlim([-0.2, 1.2])
#plt.ylim([-0.2, 0.2])

#plt.grid(True, lw=1, zorder=0)

#plt.xticks(fontsize=16)
#plt.yticks(fontsize=16)


#Set proportional aspect ratio
ax.set_aspect('equal')


#Turn off axes
plt.axis('off')


#Save figure to png - Other formats are available
#plt.savefig('fig1.png', dpi=300)
#plt.savefig('fig1.png', dpi=300, bbox_inches='tight')


plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(12,12)


#Plot cp data
ax1 = fig.add_subplot(211)
plt.plot(data1[:,0],data1[:,2])


#Axes bounds
ax1.set_xlim([0, 1])
#ax1.set_ylim([-1, 2])


#Axes labels and ticks format
plt.xlabel('Airfoil chord $c$', fontsize=18)
plt.ylabel('$c_p$', fontsize=18)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)


#Save figure to png - Other formats are available
#plt.savefig('fig1.png', dpi=300)
#plt.savefig('fig1.png', dpi=300, bbox_inches='tight')


plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(12,12)


#Create subplot 1
ax1 = fig.add_subplot(211)


#Plot CP data
plt.plot(data1[:,0],data1[:,2])


#Bounds of subplot 1
ax1.set_xlim([0, 1])
#ax1.set_ylim([-1, 2])


#Axes labels and ticks format of subplot 1
plt.xlabel('Airfoil chord $c$', fontsize=18)
plt.ylabel('$c_p$', fontsize=18)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)


#Create subplot 2
ax2 = fig.add_subplot(212)


#Plot airfoil data
plt.plot(airfoil[:,0],airfoil[:,1])


#Axes labels and ticks format of subplot 1
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)


#Set proportional aspect ratio in subplot 2
ax2.set_aspect('equal')


#Turn off axes
#plt.axis('off')


#Tigh plot - Recommended when using subplots
plt.tight_layout()


#Save figure to png - Other formats are available
#plt.savefig('fig1.png', dpi=300)
#plt.savefig('fig1.png', dpi=300, bbox_inches='tight')


plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(12,6)


#Plot in vertical axis 1 (left)
ax1 = fig.add_subplot()


#Plot in vertical axis 2 (right)
ax2 = ax1.twinx()


#Plot CP data in axis 1
ax1.plot(data1[:,0],1*data1[:,2],zorder=10)


#Axes labels and ticks format of axis 1
ax1.set_xlabel('Airfoil chord $c$', fontsize=18)
ax1.set_ylabel('$c_p$', fontsize=18)


#Plot airfoil data in axis 1
ax2.plot(airfoil[:,0],airfoil[:,1],color='black',alpha=0.5,zorder=0)


#Set proportional aspect ratio in axis 2 and set labels
ax2.set_aspect('equal')
ax2.set_ylabel('Airfoil thickness $t/c$', fontsize=18)


#Axes labels 
ax1.tick_params(axis ='x',labelsize=16)
ax1.tick_params(axis ='y',labelsize=16)
ax2.tick_params(axis ='y',labelsize=16)


#Bounds in axes
ax1.set_xlim([-0.2, 1.2])
ax1.set_ylim([-1.25,1.25])
ax2.set_ylim([-0.2,0.2])


#Plt grid using axis 1
ax1.grid()


#Save figure to png - Other formats are available
#plt.savefig('fig1.png', dpi=300)
#plt.savefig('fig1.png', dpi=300, bbox_inches='tight')


plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(12,6)


#Plot in vertical axis 1 (left)
ax1 = fig.add_subplot()


#Plot in vertical axis 2 (right)
ax2 = ax1.twinx()


#Plot CP data in axis 1
#Data was divided into top and bottom
ax1.plot(data2[:,0],1*data2[:,2],label='Top surface',zorder=10)
ax1.plot(data3[:,0],1*data3[:,2],label='Bottom surface',zorder=11)


#Axes labels and ticks format of axis 1
ax1.set_xlabel('Airfoil chord $c$', fontsize=18)
ax1.set_ylabel('$c_p$', fontsize=18)


#Plot airfoil data in axis 1
ax2.plot(airfoil[:,0],airfoil[:,1],color='black',alpha=0.5,zorder=0)


#Set proportional aspect ratio in axis 2 and set labels
ax2.set_aspect('equal')
ax2.set_ylabel('Airfoil thickness $t/c$', fontsize=18)


#Axes labels 
ax1.tick_params(axis ='x',labelsize=16)
ax1.tick_params(axis ='y',labelsize=16)
ax2.tick_params(axis ='y',labelsize=16)


#Bounds in axes
ax1.set_xlim([-0.2, 1.2])
ax1.set_ylim([-1.5,1.5])
ax2.set_ylim([-0.2,0.2])


#Invets axis 1
ax1.invert_yaxis()


#Plot legend using axis 1
ax1.legend(loc=0, fontsize=18)



#Major ticks
ax1.tick_params(axis='both', size=10, width=1, labelsize=16)

#Minor ticks
ax1.xaxis.set_minor_locator(AutoMinorLocator())
ax1.yaxis.set_minor_locator(AutoMinorLocator())
ax1.tick_params(which='minor', size=5, width=1)
ax2.yaxis.set_minor_locator(AutoMinorLocator())
ax2.tick_params(which='minor', size=5, width=1)


#Plot grid
#ax1.grid(True, lw=1, zorder=1)
ax1.grid(which='major', linestyle='-', zorder=1)
ax1.grid(which='minor', linestyle='-.', linewidth=0.5, zorder=1)


#Add annotation
plt.annotate('Minimum $c_p$',
            xy=(0.3,0.16), arrowprops=dict(arrowstyle='->'), xytext=(0.3, 0.3),fontsize=12)


#Save figure to png - Other formats are available
#plt.savefig('fig1.png', dpi=300)
#plt.savefig('fig1.png', dpi=300, bbox_inches='tight')


plt.show()
```
